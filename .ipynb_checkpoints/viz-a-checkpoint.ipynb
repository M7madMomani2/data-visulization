## Relation between total_cases and total_tests

```python
avg_total_cases = covid19[["total_cases"]].mean()
avg_total_tests= covid19[["total_tests"]].mean()
avg_total_cases=avg_total_cases.values
avg_total_tests=avg_total_tests.values
percentage_cases=avg_total_tests[0]/avg_total_cases[0]
sns.barplot(x='total_cases', y='total_deaths', data=covid19)
```
